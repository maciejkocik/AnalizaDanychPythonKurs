# <span style="color:#312e81; font-weight:600;">3. Github + GitKraken

1. Czym jest GitHub
- GitHub to internetowa platforma do hostowania kodu, współpracy zespołowej i zarządzania wersjami projektów.
- GitHub opiera się na systemie kontroli wersji Git (Git ≠ GitHub).
- Umożliwia śledzenie zmian w plikach, cofanie się do wcześniejszych wersji oraz pracę wielu osób nad tym samym projektem.

2. Czym jest Git
- Git to rozproszony system kontroli wersji (distributed version control system).
- Każda osoba ma pełną kopię historii projektu na swoim komputerze.
- Git zapisuje zmiany w postaci commitów (punktów w historii).

3. Podstawowe pojęcia
- Repozytorium (repository / repo) – folder projektu z historią zmian.
- Commit – zapis zmian z opisem co i dlaczego zostało zmienione.
- Branch (gałąź) – równoległa wersja projektu (np. do testów).
- Main / master – główna, stabilna wersja projektu.
- Clone – skopiowanie repozytorium na lokalny komputer.
- Push / Pull – wysyłanie zmian na GitHub / pobieranie zmian z GitHuba.

4. Jak wygląda typowy workflow
	1.	Tworzysz repozytorium na GitHubie.
	2.	Klonujesz je na swój komputer.
	3.	Edytujesz pliki lokalnie.
	4.	Zapisujesz zmiany jako commit.
	5.	Wypychasz (push) zmiany na GitHub.
	6.	Inni mogą:
        - pobrać zmiany (pull),
        - zgłosić poprawki,
        - komentować kod.

## <span style="color:#312e81; font-weight:600;">4. Import danych

### <span style="color:#312e81; font-weight:600;">1. Dlaczego import danych jest kluczowy?

1. Dane są punktem wyjścia każdego pipeline’u analitycznego

Każda analiza – eksploracyjna, inferencyjna czy predykcyjna – zaczyna się od poprawnego wczytania danych do środowiska obliczeniowego.
Błędy na tym etapie propagują się w dół pipeline’u (garbage in → garbage out).

2. Import to operacja transformacyjna, nie tylko „wczytanie pliku”

Import danych obejmuje:
	•	parsowanie tekstu,
	•	inferencję typów,
	•	obsługę kodowania znaków,
	•	interpretację separatorów i nagłówków,
	•	rozpoznanie braków danych (NA handling),
	•	konwersję formatów daty/czasu.

To już jest pierwszy etap przetwarzania danych (ETL – Extract, Transform, Load).

3. Reproducibility i standaryzacja

Jawne określanie:
	•	ścieżek,
	•	schematu danych,
	•	typów kolumn,
	•	parametrów parsera

jest elementem replikowalności analizy.

### <span style="color:#312e81; font-weight:600;"> 2. Po co uczyć importu w sposób świadomy (a nie „kliknij i działa”)?

1. Typy danych determinują dalsze operacje

Np.:
	•	liczby zapisane jako str → brak możliwości agregacji,
	•	daty jako object → brak operacji czasowych,
	•	zmienne kategoryczne jako int → błędne modelowanie.

Studenci powinni rozumieć:
	•	różnicę między reprezentacją w pliku a reprezentacją w pamięci,
	•	konwersję typu jako operację kosztowną obliczeniowo i semantycznie.

2. Dane realne ≠ dane „czyste”

W praktyce występują:
	•	mieszane separatory,
	•	różne kodowania (UTF-8 vs Windows-1250),
	•	nieregularne nagłówki,
	•	puste wiersze,
	•	niestandardowe oznaczenia braków (999, “.”, “-”).

3. Skalowalność

Import ma znaczenie przy:
	•	dużych plikach (memory footprint),
	•	pracy na serwerach,
	•	przetwarzaniu strumieniowym.

### <span style="color:#312e81; font-weight:600;"> 3. Jak działa import danych w Pythonie (poziom koncepcyjny)

1. Parsowanie pliku tekstowego (np. CSV)

Schemat działania:
	1.	Odczyt bajtów z dysku.
	2.	Dekodowanie (np. UTF-8).
	3.	Tokenizacja na wiersze.
	4.	Podział na kolumny wg separatora.
	5.	Inferencja typów.
	6.	Konstrukcja struktury danych (np. DataFrame).

To proces sekwencyjny i podatny na błędy formatowania.


2. Reprezentacja danych w pamięci

W pandas:
	•	dane przechowywane są kolumnowo,
	•	każda kolumna ma określony dtype,
	•	operacje są wektorowe.

W kontekście dydaktycznym warto wyjaśnić:
	•	różnicę między object, int64, float64, category, datetime64,
	•	konsekwencje dla pamięci i wydajności.


3. Import jako operacja deterministyczna vs heurystyczna

Domyślny import:
	•	korzysta z heurystyki rozpoznawania typów.

Import świadomy:
	•	explicite określa dtype,
	•	definiuje na_values,
	•	kontroluje parse_dates,
	•	ustawia encoding.

To różnica między „zadziałało” a „jest metodologicznie poprawne”.

### <span style="color:#312e81; font-weight:600;"> 4. Główne formaty danych, które studenci powinni znać

| Format           | Charakterystyka                    | Typowe zastosowanie                     |
|------------------|------------------------------------|-----------------------------------------|
| CSV              | Tekstowy, uniwersalny              | Dane tabelaryczne                       |
| Excel (.xlsx)    | Format binarny + arkusze           | Współpraca z biznesem                   |
| JSON             | Struktury zagnieżdżone             | API, dane webowe                        |
| Parquet          | Kolumnowy, skompresowany           | Duże zbiory danych                      |
| SQL              | Relacyjne bazy danych              | Dane produkcyjne                        |
